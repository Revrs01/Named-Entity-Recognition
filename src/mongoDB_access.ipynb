```python
import json
import pandas
import NER_tag_converter
import pymongo

aggregate_data = list()
password = ''
link_db = pymongo.MongoClient('mongodb+srv://philwu53:{}@clusterner.0wnlvev.mongodb.net/test'.format(password))
# load dataframe
data = pandas.read_csv('../Datasets/self_create_datasets_with_diff_tags/Train_data_ner_tags.csv', encoding='utf-8')
# convert NER token number to NER tag
for i in range(len(data['Id'])):
    index = int(data['Id'][i])
    token = eval(data['Tokens'][i])
    tag = list()
    for j in eval(data['NER_tag'][i]):
        tag.append(NER_tag_converter.convert(j))
    # create dictionary for every index
    json_format = {'index': str(index), 'token': str(token), 'NER_tag': str(tag)}
    aggregate_data.append(json_format)
    # insert into mongoDB

# access into specific table
mydb = link_db['mydatabase']
mytable = mydb['table']
# insert all data into database
mytable.insert_many(aggregate_data)
```
